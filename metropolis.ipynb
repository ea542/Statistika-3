```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.stats import norm, beta, uniform
```

```python
f = beta(5, 2)     # cilove rozdeleni
q = norm(0, 0.1)   # navrhove rozdeleni
```

```python
xs = np.linspace(0, 1, 400)
plt.plot(xs, f.pdf(xs))
```

```python
N = 10000    # pocet vzorku
x0 = 0.2     # 
samples = np.array([x0])    # generovane vzo
```

```python
for n in range(N):
    xprop = samples[-1]+q.rvs()
    A = np.min([f.pdf(xprop)/f.pdf(samples[-1]), 1])
    U = uniform.rvs()
    xnew = xprop if U<A else samples[-1]
    #print (A, xnew)
    samples = np.append(samples, xnew)
```

```python
plt.hist(samples, bins=40, density=True)
plt.plot(xs, f.pdf(xs), 'r')
```
